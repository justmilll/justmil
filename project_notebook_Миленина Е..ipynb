Создадим функцию, которая проверяет значения столбца games_sales и выводит результат в виде диапазона.

```python
def games_kategory(games_sales):
    
    if games_sales <= 6:
        return '0.0-6.0'
    if games_sales <= 12:
        return '6.1-12.0'
    if games_sales <= 20:
        return '12.1-20.0'
    if games_sales <= 30:
        return '20.1-30.0'
    return '30.1 and more' 
```

```python
def popular_games(column):

    games_kategory = column['games_kategory']
    
    if games_kategory == '30.1 and more':
        return 'popular'
    return '-'
```

```python
%%time

model = LinearRegression()
```
